```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('women-stem.csv')
df.head()
```

```python
def majorityWomen(x):
    if x > .50:
        return 'Majority Women'
    else:
        return 'Majority Men'

df['Gender Majority'] = df['ShareWomen'].apply(majorityWomen)    
```

```python
df.head()
```

# Plotting Numerical Variables

```python
sns.relplot(x='ShareWomen', y = 'Median', data = df)
plt.show()
```

```python
# add a title
sns.relplot(x='ShareWomen', y = 'Median', data = df)
plt.title('Median Salary vs Share of Women in a Major')
plt.show()
```

```python
sns.set_palette('Paired', 10)

sns.relplot(x='ShareWomen', y='Median', hue = 'Major_category', data = df)
plt.title('Median Salary vs Share of Women in a Particular Major')
plt.show()
```

```python
sns.relplot(x='ShareWomen', y='Median', hue = 'Major_category', style = 'Major_category', data = df)
plt.show()
```

```python
# using color to represent a continuous variable
sns.relplot(x='ShareWomen', y='Median', hue = 'Total', data = df)
plt.title('Median Salary vs Share of Women in a Particular Major')
plt.show()
```

```python
sns.relplot(x='ShareWomen', y='Median', hue = 'Major_category', size = 'Total', sizes = (15, 200), data = df)
plt.title('Median Salary vs Share of Women in a Particular Major')
plt.show()
```

```python
sns.relplot(x= 'ShareWomen', y = 'Women', kind = 'line', data = df)
plt.title('Share of Women vs Total Women in a Particular Major')
plt.show()
```

```python
# Can break things up based on condtion

sns.relplot(x = 'ShareWomen', y = 'Median', col = 'Gender Majority', 
            data = df)
plt.show()
```

```python
# Scatterplot with a line of best fit 
sns.lmplot(x = 'ShareWomen', y='Median', data = df)
plt.show()
```

```python
# Histogram of Median salaries 

x = df['Median']
sns.distplot(x, bins = 10)
plt.title('Distribution of Median Salaries')
plt.show()

# NOTE: can't do x = 'col_name', data = df using distplot()
```

```python
# Want distribution curve but not bins

sns.kdeplot(x)
plt.show()
```

```python
# Want shaded distribution curve

sns.kdeplot(x, shade = True)
plt.show()
```

```python
#  Relationships & distributions of the plotted variables 

sns.jointplot(x= 'ShareWomen', y = 'Median', data = df)
plt.show()
```

```python
# Linear regression line on a joint plot 

sns.jointplot(x= 'ShareWomen', y = 'Median', data = df, kind = 'reg')
plt.show()
```

```python
# pairplots to view relationship between all different numerical variables
sns.pairplot(df)
plt.show()
```

# Plotting Categorical Variables

```python
# Scatter plots on categorical variables
sns.catplot(x = 'Major_category', y = 'Median', data = df)
plt.title('Median Salary of Different Major Categories')
plt.show()
```

```python
# fix the x-axis labels - can switch the data on the x- and y- axis
sns.catplot(x = 'Median', y = 'Major_category', data = df)
plt.title('Median Salary of Different Major Categories')
plt.show()
```

```python
# second option- rotate the x-axis labels

sns.catplot(x = 'Major_category', y = 'Median', data = df)
plt.xticks(rotation = 90)
plt.title('Median Salary of Different Major Categories')
plt.show()
```

```python
# Specify which majors are majority men and majority women with color 

sns.catplot(x = 'Major_category', y = 'Median', hue = 'Gender Majority', data = df)
plt.title('Median Salary of Different Major Categories')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Adjust the alignment of the points

sns.catplot(x = 'Major_category', y = 'Median', jitter = False, data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
# try a different alignment

sns.catplot(x = 'Major_category', y = 'Median', kind = 'swarm', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
# View distribution of data points with boxplots 

sns.catplot(x = 'Major_category', y = 'Median', kind = 'box', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
# use a boxen plot to better view the distribution within the boxplots and
# identify if there are heavy tails in the distribution

sns.catplot(x = 'Major_category', y = 'Median', kind = 'boxen', data = df)
plt.title('Boxen Plot')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Use a violin plot to better view distribution of salaries 

sns.catplot(x = 'Major_category', y = 'Median', kind = 'violin', data = df)
plt.title('Violin Plot')
plt.xticks(rotation = 90)
plt.show()
```

```python
#split the violin plots based on which gender has majority in specific major 

sns.catplot(x = 'Major_category', y = 'Median', kind = 'violin', 
            hue = 'Gender Majority', split = 'True', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
# another way to view distribution is with a swarm plot on top of the violin plot

sns.catplot(x = 'Major_category', y = 'Median', kind = 'violin', data = df)
sns.swarmplot(x = 'Major_category', y = 'Median', data = df, color = 'black')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Create a bar graph to view the number of majors in each category

sns.catplot(x = 'Major_category', kind = 'count', data = df)
plt.xticks(rotation = 90)
plt.show()
```

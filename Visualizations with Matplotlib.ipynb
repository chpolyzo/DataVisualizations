```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WorldHappiness2018_Data.csv')
df.head()
```

```python
# we can represent the relationship between rank and score with a line graph, as we'd expect it to be linear- the 
# higher the scorel, the lower (closer to 1) the rank 

rank = df['Rank']
score = df['Score']

plt.plot(rank, score)
plt.show()
```

We should add a title and some axes labels so we know what this graph is showing us

```python
plt.plot(rank, score)
plt.title('Happiness Score vs World Rank')
plt.xlabel('Country Rank')
plt.ylabel('Country Score')
plt.show()
```

What happens if we want to see more than one relationship on the same graph?

```python
gdp = df['GDP_Per_Capita']

plt.plot(rank, score)
plt.plot(rank, gdp)
plt.show()
```

There's a problem with this: the values that the two variables hold are on different scales, so it's hard to get the whole story just by looking at this.

```python
gdp = df['GDP_Per_Capita']
lifeExp = df['Healthy_Life_Expectancy']

plt.plot(rank, gdp)
plt.plot(rank, lifeExp)
plt.show()
```

```python
# but what color is which? 
plt.plot(rank, gdp)
plt.plot(rank, lifeExp)
plt.title('World Rank vs GDP and Life Expectancy')
plt.xlabel('Country Rank')
plt.legend()
plt.show()
```

```python
# what if we want to change the colors?
plt.plot(rank, gdp, color = 'green')
plt.plot(rank, lifeExp, color = 'blue')
plt.title('World Rank vs GDP and Life Expectancy')
plt.xlabel('Country Rank')
plt.legend()
plt.show()
```

```python
# What if we want to see if the relationship is good for linear regression? 
plt.scatter(gdp, score)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# Convention: axes should start from 0 

plt.scatter(gdp, score)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.ylim(0,8)
plt.show()
```

```python
# Check out the distributions of some of the variables  

plt.hist(score)
plt.title('Happiness Score Distribution')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```
